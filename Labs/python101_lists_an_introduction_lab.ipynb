## üí° Quick Reference Guide

**List Methods You Learned:**
- `list.append(item)` - Add to the end
- `list.insert(position, item)` - Add at specific position
- `list.pop(position)` - Remove and return item at position
- `list.remove(value)` - Remove first occurrence of value
- `list.reverse()` - Flip the list backwards
- `list.sort()` - Arrange in order (smallest to largest)
- `list.index(item)` - Find position of item
- `list.count(item)` - Count how many times item appears

**Remember:**
- Python counts from 0! (First item is at position 0)
- List methods use dot notation: `list_name.method()`
- Built-in functions use: `function_name(data)`
