### 💡 Reminders Before You Start

- **Mixed list** → holds different data types in one list: `['Anna', 15, True, 3.9]`
- **Nested list** → a list that contains other lists: `['Anna', 15, ['math', 'PE']]`

- Use `list[index]` to access items (index starts at **0**)
- Use `list[i][j]` to access items inside a nested or 2D list
- Use `len(list)` for number of rows, `len(list[0])` for number of columns
