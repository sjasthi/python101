### 💡 Reminders Before You Start

- Strings are **sequences of characters** — they support indexing just like lists: `name[0]` gives the first character
- Indexing starts at **0**. Use **negative indexing** to count from the end: `name[-1]` gives the last character
- **Slicing** uses `string[start:stop:step]` — the `stop` index is **not** included
- Strings are **immutable** — you cannot change individual characters: `name[0] = 'C'` causes a TypeError
- All string methods return **new strings** — they do **not** modify the original
- The `in` and `not in` operators are **case-sensitive**: `'python' in 'Python'` returns `False`
- `split()` breaks a string into a **list**; `join()` glues a list back into a **string**
