## Exercise 6: Using Finally (1 point)

**Task:** Add a finally block that always prints "Thank you for using the calculator!"

**Hint:** The finally block runs no matter what - whether an exception occurs or not.

```python
# Add a finally block to this code

try:
    x = int(input("Enter a number: "))
    result = 100 / x
    print(f"100 divided by {x} is {result}")
except ValueError:
    print("Please enter a valid number!")
except ZeroDivisionError:
    print("Cannot divide by zero!")
else:
    print("Calculation completed successfully!")
# YOUR CODE HERE - Add finally block
```
