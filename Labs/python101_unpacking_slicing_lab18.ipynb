Remember:
- **Unpacking** = Extract tuple values: `name, score, level = player_tuple`
- **Basic Slice** = `list[start:end]` - end is exclusive!
- **Negative Index** = Count from end: `list[-1]` is last item
- **Step Slice** = `list[start:end:step]` - skip items
- **Slice Shortcuts**: `[:3]` = first 3, `[3:]` = from 3 to end, `[::2]` = every other
