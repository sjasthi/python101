### 💡 Reminders Before You Start

- A **tuple** uses **round brackets** `( )` — not square brackets: `planets = ("Mercury", "Venus", "Earth")`
- Tuples are **immutable** — you cannot change, add, or remove items after creation
- Indexing works exactly like lists — index starts at **0**, and `-1` gives the last item
- Only **two methods** work on tuples: `index(x)` and `count(x)`
- Built-in functions like `len()`, `min()`, `max()`, `sum()`, `sorted()` all work on tuples
- To create a **single-element tuple**, you must add a trailing comma: `(42,)` ← the comma is required!
