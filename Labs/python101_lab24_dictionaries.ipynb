### 💡 Reminders Before You Start

- A **dictionary** uses **curly braces with colons** `{ key: value }` — example: `student = {"name": "Alice", "age": 14}`
- To create an **empty dictionary**, use `{}` or `dict()`
- Access values using the **key**: `student["name"]` returns `"Alice"`
- Use `.get(key)` for **safe access** — returns `None` instead of crashing if the key is missing
- **Keys must be unique** — if you assign the same key twice, the last value wins
- Add or update: `student["grade"] = "8th"` — adds if new, updates if existing
- Delete: `del dict[key]`, `.pop(key)`, `.popitem()`, or `.clear()`
- Loop helpers: `.keys()`, `.values()`, `.items()`
- The `in` keyword checks **keys**, not values

### ⚠️ Common Pitfalls

| Pitfall | What Happens | Fix |
|---|---|---|
| Using `dict[0]` to access first item | `KeyError: 0` (unless `0` is an actual key) | Use the actual key: `dict["name"]` |
| Forgetting quotes around string keys | `NameError: name 'age' is not defined` | Always use quotes: `dict["age"]` |
| Duplicate keys in definition | Last value silently overwrites the first | Each key should be unique |
| Using `in` to check values | `in` checks **keys** only | Use `value in dict.values()` |
| Confusing `[]` vs `.get()` | `[]` crashes with KeyError if key missing | Use `.get()` for safe access |
