### 💡 Reminders Before You Start

- A **set** uses **curly braces** `{ }` — example: `fruits = {"apple", "banana", "cherry"}`
- To create an **empty set**, use `set()` — NOT `{}` (that creates an empty dictionary!)
- Sets are **unordered** and **unindexed** — you **cannot** use `[0]`, `[1]`, etc.
- Sets **do not allow duplicates** — Python automatically removes them
- Only **immutable** items (numbers, strings, booleans, tuples) can go inside a set — **not lists**
- Use `add()` for one item, `update()` for multiple items
- Use `remove()` (raises error if missing) or `discard()` (no error if missing)
- Set operations: `|` or `.union()`, `&` or `.intersection()`, `-` or `.difference()`, `^` or `.symmetric_difference()`

### ⚠️ Common Pitfalls

| Pitfall | What Happens | Fix |
|---|---|---|
| Using `{}` for empty set | Creates a dictionary instead! | Use `set()` |
| Using `[0]` on a set | `TypeError: 'set' object is not subscriptable` | Sets don't support indexing |
| Expecting a fixed order | Sets are unordered — output order may vary | That's normal! |
| Confusing `-` and `^` | `-` = difference (one-sided), `^` = symmetric difference (both sides) | Think: `-` is one-way, `^` is both-ways |
