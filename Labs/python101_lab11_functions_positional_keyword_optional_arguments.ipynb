## üìö What You'll Learn

In this lab, you'll experiment with three types of function arguments:

1. **Positional Arguments** - Arguments that must be in the correct order
2. **Keyword Arguments** - Arguments you call by name (order doesn't matter!)
3. **Optional Arguments** - Arguments that have default values (you can skip them!)

### Quick Example:
```python
def greet(name, greeting="Hello", excited=False):
    message = f"{greeting} {name}"
    if excited:
        message += "!!!"
    return message

# Different ways to call it:
greet("Alice")                          # Uses default greeting
greet("Bob", "Hi")                      # Changes greeting
greet("Charlie", excited=True)          # Uses keyword argument
greet(greeting="Hey", name="Dana")      # All keywords (order doesn't matter!)
```
