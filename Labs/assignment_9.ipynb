### ⚠️ Common Pitfalls

| Pitfall | What Happens | Fix |
|---|---|---|
| Forgetting `.strip()` when reading lines | Extra whitespace causes mismatches | Always `.strip()` each line |
| Forgetting `.lower()` in Problem 6 | "Python" and "python" treated as different words | Convert to lowercase first |
| Not skipping 1904 and 1994 in Problem 7 | Years and teams get misaligned | Add `if` checks in your year loop |
| Using `=` instead of `+=` when counting | Resets count to 1 every time | Check if key exists first, then `+= 1` |
| Forgetting `int()` on `input()` | Year stored as string, not integer | Wrap in `int()`: `int(input(...))` |
