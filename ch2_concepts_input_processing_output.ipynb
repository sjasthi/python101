# [6] Magic Numbers: Avoid those

````python
#@title Avoid Magic Numbers in the code!

# What is this 100? What is this 9.8?
# What are these magic numbers?
# bad programming practice
interest = 100 *  0.08


# Good programming practice
PRINCIPLE = 100
INTEREST_RATE = 0.08

interest = PRINCIPLE * INTEREST_RATE

#print(interest)
print('Interest earned: ', interest)
````

````python
#@title Avoid Magic Numbers in the output

# magic numbers in the output is a bad practice
print(interest)
print(interest + PRINCIPLE)


# good practice
print('Interest is: ', interest)
print('Principle after interest: ', interest + PRINCIPLE)
````

# [7] Breaking Long Lines using \ or ()

````python
#@title Breaking long lines into smaller lines (using \ line break)

value_1 = 1
value_2 = 2
value_3 = 3
value_4 = -1
value_5 = 0
value_6 = -2

# Using long lines
total = value_1 + value_2 + value_3 + value_4 + value_5 + value_6
print('Total: ', total)

# Using \ line break
total = value_1 + value_2 + value_3 + \
        value_4 + value_5 + value_6
print('Total: ', total)

````

````python
#@title Breaking long lines into smaller lines (using () paranthesis)

value_1 = 1
value_2 = 2
value_3 = 3
value_4 = -1
value_5 = 0
value_6 = -2
total = value_1 + value_2 + value_3 + value_4 + value_5 + value_6
print('Total: ', total)

# Using a paranthesis
total = (value_1 + value_2 + value_3 +
        value_4 + value_5 + value_6)
print('Total: ', total)
````
