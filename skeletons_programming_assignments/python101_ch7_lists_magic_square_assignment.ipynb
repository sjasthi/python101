## 📖 Background: What is a Lo Shu Magic Square?

The **Lo Shu Magic Square** is one of the oldest math puzzles in history — it comes from ancient China and is over **4,000 years old**! 🐉

It is a **3×3 grid** (3 rows, 3 columns) with these two special properties:

### Property 1 — Contains numbers 1 through 9 exactly once
Every number from **1 to 9** appears in the grid. No repeats, no missing numbers!

### Property 2 — All rows, columns, and diagonals sum to 15
When you add up any row, any column, or either diagonal, you **always get 15**.

```
  ┌───┬───┬───┐
  │ 2 │ 7 │ 6 │  → 2 + 7 + 6  = 15  ✓
  ├───┼───┼───┤
  │ 9 │ 5 │ 1 │  → 9 + 5 + 1  = 15  ✓
  ├───┼───┼───┤
  │ 4 │ 3 │ 8 │  → 4 + 3 + 8  = 15  ✓
  └───┴───┴───┘
     ↓   ↓   ↓
    15  15  15  ← column sums ✓

  Diagonal 1 (top-left → bottom-right): 2 + 5 + 8 = 15  ✓
  Diagonal 2 (top-right → bottom-left): 6 + 5 + 4 = 15  ✓
```

The magic sum for a 3×3 grid is always **15**.

### Example 1: Representing a 3×3 Grid as a 2D List

````python
# A 3x3 grid stored as a 2D list
magic = [
    [2, 7, 6],   # Row 0
    [9, 5, 1],   # Row 1
    [4, 3, 8]    # Row 2
]

print("The grid:")
for row in magic:
    print(row)

print()
print("Number of rows:   ", len(magic))
print("Number of columns:", len(magic[0]))
print("Top-left cell:    ", magic[0][0])   # row 0, col 0
print("Center cell:      ", magic[1][1])   # row 1, col 1
print("Bottom-right:     ", magic[2][2])   # row 2, col 2
````

### Example 2: Summing a Row

````python
magic = [
    [2, 7, 6],
    [9, 5, 1],
    [4, 3, 8]
]

# Sum each row using a loop
print("Row sums:")
for i in range(len(magic)):          # i = 0, 1, 2  (row index)
    row_sum = 0
    for j in range(len(magic[0])):   # j = 0, 1, 2  (column index)
        row_sum = row_sum + magic[i][j]
    print(f"  Row {i}: {magic[i]}  → sum = {row_sum}")

# Python shortcut: sum() works on a list too!
print()
print("Using sum() shortcut:")
for i in range(len(magic)):
    print(f"  Row {i} sum = {sum(magic[i])}")
````

### Example 3: Summing a Column

````python
magic = [
    [2, 7, 6],
    [9, 5, 1],
    [4, 3, 8]
]

# Column-major: outer loop = columns, inner loop = rows
print("Column sums:")
for j in range(len(magic[0])):    # j = 0, 1, 2  (column index)
    col_sum = 0
    for i in range(len(magic)):   # i = 0, 1, 2  (row index)
        col_sum = col_sum + magic[i][j]
    print(f"  Column {j} sum = {col_sum}")
````

### Example 4: The Two Diagonals

````python
magic = [
    [2, 7, 6],
    [9, 5, 1],
    [4, 3, 8]
]

# Diagonal 1: top-left to bottom-right → grid[0][0], grid[1][1], grid[2][2]
# Notice: row index == column index (both are i)
diag1 = 0
for i in range(len(magic)):
    diag1 = diag1 + magic[i][i]
print("Diagonal 1 (top-left → bottom-right):", diag1)

# Diagonal 2: top-right to bottom-left → grid[0][2], grid[1][1], grid[2][0]
# Notice: row index + column index = 2  (which is len-1)
diag2 = 0
n = len(magic)          # n = 3
for i in range(n):
    diag2 = diag2 + magic[i][n - 1 - i]
print("Diagonal 2 (top-right → bottom-left):", diag2)
````

### Example 5: Checking Whether Numbers 1–9 Are Present

````python
magic = [
    [2, 7, 6],
    [9, 5, 1],
    [4, 3, 8]
]

# Step 1: Collect all 9 values into a flat list using row-major traversal
flat = []
for row in magic:
    for num in row:
        flat.append(num)

print("Flat list of all values:", flat)

# Step 2: Sort the flat list
print("Sorted:                 ", sorted(flat))

# Step 3: Compare to [1, 2, 3, 4, 5, 6, 7, 8, 9]
expected = list(range(1, 10))    # [1, 2, 3, 4, 5, 6, 7, 8, 9]
print("Expected:               ", expected)
print("Contains 1-9 exactly?   ", sorted(flat) == expected)
````
