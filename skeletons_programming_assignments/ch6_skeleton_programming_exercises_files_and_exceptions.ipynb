````python
#@title 12.2.Average Steps Taken  (List Based Implementation)
# Global Variables/Constants


# Maintain a tuple of month names
months = ('Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul',
         'Aug', 'Sep', 'Oct', 'Nov', 'Dec')

# Maintain a tuple of 'count of days' for each month
days = (31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31)

# Open the steps.txt file for reading
steps_file_obj = open('steps.txt', 'r')

# initialize the month.
month_number = 0

# This loops through the months
for num in range(1, 13):
    # Reset base values for the counters for each month
    total_steps = 0
    count = 0
    average = 0

    # This iterates until the end of the month is reached
    for count in range(1, days[month_number] + 1):
        # This reads the lines until the end of the month is reached
        steps = int(steps_file_obj.readline())

        # This sums the months
        total_steps = total_steps + steps

    # This averages the steps for the month
    average = total_steps / days[month_number]

    # This prints the average for the month
    print (months[month_number] + '\t\t' + str(total_steps) + '\t\t' + format(average, ',.2f'))

    # Increment the month number
    month_number = month_number + 1

# close the file
steps_file_obj.close()
````

````python
#@title 12.4.Average Steps Taken  (List Based Implementation - Using slicing)
# Maintain a tuple of month names
months = ('Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec')

# Maintain a tuple of 'count of days' for each month
days = (31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31)

# Open the steps.txt file for reading
steps_file_obj = open('steps.txt', 'r')

# Read all the lines into one big list
steps_list = steps_file_obj.readlines()
steps_list = [int(x) for x in steps_list]
print("count of days: ", len(steps_list))

start_index = 0
end_index = 0
for month_number in range(0,12):
  start_index = end_index
  end_index = end_index + days[month_number]
  month_list = steps_list[start_index:end_index]
  total = sum(month_list)
  average = total / len(month_list)
  print (months[month_number] + '\t\t' + str(total) + '\t\t' + format(average, ',.2f'))

# close the file
steps_file_obj.close()

````

````python
file_name = 'steps.txt'
days_in_month = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

with open(file_name, 'r') as file_object:
    line_number = 0
    total_steps = 0

    print("Month\t\tTotal Steps\t\tAverage")

    for month, days in enumerate(days_in_month, start=1):
        for _ in range(days):
            line_number += 1
            current_steps = int(file_object.readline())
            total_steps += current_steps

        average = total_steps / days
        print(f"{month:02d}\t\t{total_steps}\t\t{average:.2f}")
        total_steps = 0
````
