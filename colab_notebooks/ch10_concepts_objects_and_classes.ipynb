````python
#@title Bank_Account class
'''
You will be creating a BankAccount class in this example.

Assume that it has the following five variables.

* Name of the Account Holder
* Type of Bank Account (Savings or Checking)
* Current Balance
* Withdrawl Limit  (amount that can be withdrawn at a given time)
* Interest Rate (Some interest is paid on the Savings account balances only; And the interest is added back to the current balance of the amount)

And it has the following methods.

__init__ method (constructor)
__repr__ method (to support print method)
get methods (to return individual variables/properties)

deposit(x): this method deposits x dollars into the account; The amount deposited can not be negative.

withdraw(x): this method withdraws x dollars from the account; x can not be negative; If the x is more than the withdrawl limit, then only 'withdrawl limit' dollars are given to the customer. If there is not enough money in the account, only the available amount is given to the customer.

add_interest(days): this method calculates the interest only if the account is of type "Savings".  It calculates the interest for the given "days" and adds the interest to the current balance.

# create two instances (one savings account and one checking account)
# call deposit and withdraw methods
# ensure that your output is informative.


# Draw the UML diagram for the bank_account class.
# Refer to Chapter 10 of the text book to understand UML
'''

class Bank_Account:
  def __init__(self, name, type, balance, withdraw_limit, intrest_rate):
    self.name = name
    self.type = type
    self.balance = balance
    self.withdraw_limit = withdraw_limit
    self.intrest_rate = intrest_rate / 1000

  def __repr__(self):
    ''' String representation of Bank Account Object '''
    bank_details_str = ('\nAccount ownder: ' + self.name +
                       '\nType of account: ' + self.type +
                       '\nBalance: ' + str(self.balance) +
                       '\nWithdraw limit: ' + str(self.withdraw_limit) +
                       '\nIntrest rate: ' + str(self.intrest_rate) + '%')
    return bank_details_str

  def deposit(self, amount):
    ''' Deposits the amount and updates the balance if the amount is positive'''
    if amount <= 0:
      print("Can't deposit that little!")
      return
    self.balance += amount
    print('\nSuccessfully deposited $' + str(amount))
    print('Your new balance $' + str(self.balance))

  def withdraw(self, amount):
    ''' This method withdraws the amount and updates the balance
        One can not withdraw more than daily limit.
        Also, one can not withdraw more than the available balance
    '''
    if amount <= 0:
      print("Can't withdraw that little!")
    elif amount > self.withdraw_limit:
      amount = self.withdraw_limit
      print('\nYou tried to withdraw more than allowed! You can only withdraw up to $' + str(amount))
    elif amount >= self.balance:
      print('\nNot enough money to withdraw in the account! You withdrew $' + str(self.balance))
      self.balance = 0
    else:
      print('\nYou withdrew $' + str(amount))
      self.balance -= amount
      print('Your new balance $' + str(self.balance))

  def add_interest(self, days):
    for i in range(0, days):
      self.balance = self.balance + self.balance * self.intrest_rate
    print('\nYou will have $' + str(self.balance), 'after', str(days), 'days!')


# Create some objects and call the methods
bank_act_1 = Bank_Account('Home Financing', 'Checking', 10000, 2000, 40)
print(bank_act_1)

bank_act_1.deposit(999)
print(bank_act_1)

bank_act_1.withdraw(1000)
print(bank_act_1)
````

````python
#@title Fraction class

## EXAMPLE: simple class to represent fractions
## Added functionality by implementing +, -, *, / operators
class Fraction(object):
    """ A number represented as a fraction """
    def __init__(self, num, denom):
        """ num and denom are integers """
        self.num = num
        self.denom = denom
    def __str__(self):
        """ Returns a string representation of self """
        return str(self.num) + "/" + str(self.denom)
    def __mul__(self, other):
        """ Returns a new fraction representing the addition """
        top = self.num*other.num
        bottom = self.denom*other.denom
        return Fraction(top, bottom)
    def __add__(self, other):
        """ Returns a new fraction representing the addition """
        top = self.num*other.denom + self.denom*other.num
        bottom = self.denom*other.denom
        return Fraction(top, bottom)
    def __sub__(self, other):
        """ Returns a new fraction representing the subtraction """
        top = self.num*other.denom - self.denom*other.num
        bottom = self.denom*other.denom
        return Fraction(top, bottom)
    def __truediv__(self, other):
        """ Returns a new fraction representing the subtraction """
        top = self.num*other.denom
        bottom = self.denom*other.num
        return Fraction(top, bottom)
    def __float__(self):
        """ Returns a float value of the fraction """
        return self.num/self.denom
    def reduce(self):
        """ Returns a new fraction the reduced version of self
            using the greatest common divisor """
        def gcd(n, d):
            while d != 0:
                (d, n) = (n%d, d)
            return n
        if self.denom == 0:
            return None
        elif self.denom == 1:
            return self.num
        else:
            greatest_common_divisor = gcd(self.num,self.denom)
            top = int(self.num/greatest_common_divisor)
            bottom = int(self.denom/greatest_common_divisor)
            return Fraction(top, bottom)
    def invert(self):
        """ Returns a new fraction representing 1/self """
        return Fraction(self.denom, self.num)

# # Using shorthand operations on fractions
a = Fraction(1,4)
b = Fraction(3,4)
print(a)
c = a * b # c is a Fraction object
print(c)

# The next 3 lines are equivalent
print(a * b)                  ##1 (shorthand) same as #2, #3
print(a.__mul__(b))           ##2 (method call) same as #1, #3
print(Fraction.__mul__(a, b)) ##3 (explicit class call) same as #1, #2
########

# # The next 3 lines are equivalent
# print(float(c))              ##1 (shorthand) same as #2, #3
# print(c.__float__())         ##2 (method call) same as #1, #3
# print(Fraction.__float__(c)) ##3 (explicit class call) same as #1, #2

# # Reducing fractions
# a = Fraction(1,4)
# b = Fraction(2,3)
# c = a * b
# print(c)
# print(c.reduce())

# # Can't multiply int and Fraction
# a = Fraction(4,1)
# b = Fraction(3,9)
# ar = a.reduce()
# br = b.reduce()
# print(ar, type(ar))
# print(br, type(br))
# # c = ar * br   # gives an error bc it's multiplying an int with a Fraction


############## YOU TRY IT #####################
# Modify the str method to represent the Fraction as just the
# numerator, when the denominator is 1. Otherwise its representation
# is the numerator then a / then the denominator, as before
class Fraction(object):
    """ A number represented as a fraction """
    def __init__(self, num, denom):
        """ num and denom are integers """
        self.num = num
        self.denom = denom
    def __str__(self):
        """ Returns a string representation of self """
        # modify this
        return str(self.num) + "/" + str(self.denom)


a = Fraction(1,4)
b = Fraction(3,1)
# print(a)     # prints 1/4
# print(b)     # prints 3
````

# What is a "Class Method"? Why do we need it?


A class method in Python is a method that is bound to the class rather than the instance of the class. It is defined using the @classmethod decorator and takes the class itself as the first argument (often named cls by convention) instead of the instance (self).

We need class methods for several reasons:

**Accessing or modifying class-level variables**: Class methods can access and modify class variables. These variables are shared among all instances of the class, so class methods provide a way to manage them effectively.

**Alternative constructors**: Class methods can be used as alternative constructors. They provide a way to create new instances of a class with different initialization parameters or from different sources.

**Encapsulation and organization**: Class methods help in organizing code by grouping related functionality within the class. They encapsulate operations that are closely associated with the class itself rather than with individual instances.

**Bound to the class**: Unlike instance methods, class methods are bound to the class itself rather than to specific instances. This makes them useful for operations that don't depend on the state of a particular instance.


````python
#@title Student class with class methods
class Student:
    instructor_name = "John Doe"

    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __str__(self):
        return f"Student: {self.name}, Age: {self.age}, Instructor: {self.instructor_name}"

    @classmethod
    def change_instructor(cls, new_instructor):
        cls.instructor_name = new_instructor

    @classmethod
    def get_instructor(cls):
        return cls.instructor_name


print(Student.instructor_name)


# Create instances of Student class
student1 = Student("Alice", 20)
student2 = Student("Bob", 21)

# Display initial information
print(student1)
print(student2)

# Change instructor name using class method
Student.change_instructor("Jane Smith")

# Display updated information
print(student1)
print(student2)

# Get current instructor using class method
current_instructor = Student.get_instructor()
print("Current Instructor:", current_instructor)

````

````python
#@title @property decorator

'''
In Python, @property is a built-in decorator that allows defining methods
that can be accessed like attributes.

When a method is decorated with @property, it is called a property method.
'''

class Circle:
    def __init__(self, radius):
        self.radius = radius

    # You can only access diameter as a property. This can not be called as a method
    @property
    def diameter(self):
        return self.radius * 2

# Creating an instance of Circle
circle = Circle(5)

# Accessing the diameter property
#print(circle.diameter)  # Output: 10

# Accessign the diameter as a method
# This will not work becuase of @property decorator
# If you want to use this as a method, you should remove the @property decorator
#print(circle.diameter())
````

````python
#@title @abstractmethod decorator

'''
The @abstractmethod decorator is used in Python to define abstract methods within abstract base classes (ABCs).
An abstract method is a method declared in a class definition that does not contain an implementation.
Instead, it serves as a placeholder, and its implementation must be provided by subclasses.
'''
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimeter(self):
        pass

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

    def perimeter(self):
        return 2 * (self.width + self.height)

# Creating an instance of Rectangle
rectangle = Rectangle(5, 4)

# Calling methods
print("Area:", rectangle.area())
print("Perimeter:", rectangle.perimeter())
````

````python
#@title @classmethod decorator

#@title Student class with class methods
class Student:
    instructor_name = "John Doe"

    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __str__(self):
        return f"Student: {self.name}, Age: {self.age}, Instructor: {self.instructor_name}"

    @classmethod
    def change_instructor(cls, new_instructor):
        cls.instructor_name = new_instructor

    @classmethod
    def get_instructor(cls):
        return cls.instructor_name

# Create instances of Student class
student1 = Student("Alice", 20)
student2 = Student("Bob", 21)

# Display initial information
print(student1)
print(student2)

# Change instructor name using class method
Student.change_instructor("Jane Smith")

# Display updated information
print(student1)
print(student2)

# Get current instructor using class method
current_instructor = Student.get_instructor()
print("Current Instructor:", current_instructor)
````

### When to user staticmethod vs classmethod?

To decide whether to use @staticmethod or @classmethod you have to look inside your method.

If your method accesses other variables/methods in your class then use @classmethod.

On the other hand, if your method does not touches any other parts of the class then use @staticmethod.


From the perspective of a consumer or caller of the class methods, @staticmethod and @classmethod can often be used interchangeably. Both types of methods can be called on the class itself or on an instance of the class.

However, the choice between @staticmethod and @classmethod depends on the intended behavior and usage of the method within the class:

Use @staticmethod when:

-The method does not depend on instance or class attributes.
-The method does not need access to the instance or class itself.
-The method behaves like a utility function that logically belongs within the class namespace.


Use @classmethod when:

-The method needs access to or modification of class-level attributes or behavior.
-The method requires access to the class itself (cls) to perform some operation.
-The method is related to the class itself rather than individual instances.

In many cases, both decorators can achieve similar functionality, especially when the method does not rely on instance or class attributes. However, using the appropriate decorator helps in conveying the intended behavior and usage of the method to other developers reading the code. Additionally, using @classmethod provides more flexibility if the method requires access to class-level attributes or behavior in the future.

````python
#@title Python Class Generator (Version 2) (generates Setters and Getters for each variable)
# Function to generate the __init__ method code
# Demonstrated in class
def get_init_code(property_list):
    init_code = "\n\n# Defining the __init__ method"
    init_code += f'\n\tdef __init__(self, {", ".join(property_list)}):'

    for prop in property_list:
        init_code += f'\n\t\tself.{prop} = {prop}'

    init_code += '\n'
    return init_code

# Function to generate the __repr__ method code
def get_repr_code(class_name, property_list):

    # Start the function
    repr_code = "\n\n# Defining the __repr__ method"
    repr_code += f'\n\tdef __repr__(self):'
    repr_code += f'\n\t\treturn f"{class_name}('

    # loop through each of the properties
    for prop in property_list:
        repr_code += f'{prop}={{self.{prop}}}, '

    # remove the extra comma and space, end the parenthesis
    repr_code = repr_code[:-2]
    repr_code += ')"'

    repr_code += '\n'
    return repr_code

# Function to generate the __str__ method code
def get_str_code(class_name, property_list):

    # start the function
    str_code = "\n\n# Defining the __str__ method"
    str_code += f'\n\tdef __str__(self):'
    str_code += f'\n\t\treturn f"\\n{class_name}:'

    # loop through each of the properties
    for prop in property_list:
        str_code += f'\\n{prop}: {{self.{prop}}}'

    str_code += '"'

    str_code += '\n'
    return str_code

# Function to generate object creation code
def get_obj_creation_code(class_name, property_list):
  init_code = "\n\n# Defining the __get_obj_creation_code__ method"
  init_code += '\nsampleAddress= ' +class_name+'('
  for each_property in property_list:
    init_code += '"' + each_property + '", '
  init_code = init_code[:-2]
  init_code += ')\n'

  init_code += "print('Representation: ', repr(sampleAddress))\n"
  init_code += "print('String representation: ', str(sampleAddress))"

  return init_code


# Function to create the Python class
def create_python_class(class_name, property_list):
    python_code = f"#@title {class_name} {property_list}"
    python_code += f"\nclass {class_name}:"
    python_code += get_init_code(property_list)
    python_code += get_repr_code(class_name, property_list)
    python_code += get_str_code(class_name, property_list)
    python_code +=  get_obj_creation_code(class_name, sample_property_list)
    return python_code

# Set the class name and property list
class_name = "Book"
property_list = ["title", "author", "publisher"]
sample_property_list = ["Starting with Python", "Tony", "Pearson"]

class_name = "Beer"
property_list = ["brand", "volume", "flavor"]
sample_property_list = ["Taj", "1.5L", "dark"]


class_name = "Bread"
property_list = ["grain", "size", "shape"]
sample_property_list = ["wheat", "12", "round"]

class_name = "Ball"
propety_list = ["radius", "color", "owner"]
sample_property_list = ["4.5", "yellow", "john"]

# Create the Python class code
final_python_code = create_python_class(class_name, property_list)

# Print the generated code to the console
print(final_python_code)

# Write the code to a file
output_file_name = class_name + ".py"
output_file_obj = open(output_file_name, 'w')
output_file_obj.write(final_python_code)
output_file_obj.close()
````
