# Exceptions

-The try block lets you test a block of code for errors.

-The except block lets you handle the error.

-The else block lets you execute code when there is no error.

-The finally block lets you execute code, regardless of the result of the try- and except blocks.

## Examples of how things can go wrong while taking the inupts from the user.

Use case: Get the age from the user
====================================

Enter your age:  10   (correct input)

Enter your age:  A    (incorrect input)

Enter your age: 0     (incorrect input; age can not be 0)

Enter your age: -20   (incorrect input; age can not be negative)

Enter your age: 140   (incorrect input; no one lives for 140)

```python
#@title Version 1: Take 'age' as input from the user (1 < age < 140)
# Handle the exceptions for incorrect inputs

try:
    age = input('Enter your age: ')
    age = int(age)
    age_2 = age + 5
    print('You are now', age, 'years old')
    print('You will be', age_2, 'after 5 years')
except:
    print('Error: Something has gone wrong.')

# Test Case 1:  age = 10   (happy path; positive case)
# PASSED

# Test Case 2:  age = abc   (negative case)
# FAILED / CRASHED

```

```python
#@title Version 2: Take 'age' as input from the user (1 < age < 140)
# Handle the exceptions for incorrect inputs
try:
    age = input('Enter your age: ')
    age = int(age)
    age_2 = age + 5
    print('You are now', age, 'years old')
    print('You will be', age_2, 'after 5 years')
except ValueError:
    print('Value Error: Your input can not be converted to an integer.')

# Test Case 1:  age = 10   (happy path; positive case)
# PASSED

# Test Case 2:  age = abc   (negative case)
# PASSED

```

```python
#@title Version 3: Take 'age' as input from the user (1 < age < 140)
# Handle the exceptions for incorrect inputs
class NegativeAgeException(Exception):
  pass

class TooBigAgeException(Exception):
  pass

try:
    age = input('Enter your age: ')
    age = int(age)

    if (age < 0):
      raise NegativeAgeException('Error: Age can not be negative')

    if (age > 140):
      raise TooBigAgeException('Error: Age can not be more than 140')

    age_2 = age + 5
    print('You are now', age, 'years old')
    print('You will be', age_2, 'after 5 years')

except ValueError:
    print('Value Error: Your input can not be converted to an integer.')

except NegativeAgeException as argument:
    print(argument)

except TooBigAgeException as argument:
    print(argument)


# Test Case 1:  age = 10   (happy path; positive case)
# PASSED

# Test Case 2:  age = abc   (negative case)
# PASSED

# Test Case 3: age = -10 (negative case)
# PASSED

# Test Case 4: age = 150 (negative case)
# PASSED
```

```python
#@title Version 4: Take 'age' as input from the user (1 < age < 140)
# Given an option to the user to fix his/her mistake

# Handle the exceptions for incorrect inputs
class NegativeAgeException(Exception):
  pass

class TooBigAgeException(Exception):
  pass

# Keep on running this loop until we get the correct integer/age

while True:
      try:
          age = input('Enter your age: ')
          age = int(age)

          if (age < 0):
            raise NegativeAgeException('Error: Age can not be negative')

          if (age > 140):
            raise TooBigAgeException('Error: Age can not be more than 140')

          age_2 = age + 5
          print('You are now', age, 'years old')
          print('You will be', age_2, 'after 5 years')
          break  #we are in happy path now


      except ValueError:
          print('Value Error: Your input can not be converted to an integer.')
          print('Please try again')

      except NegativeAgeException as argument:
          print(argument)
          print('Please try again')


      except TooBigAgeException as argument:
          print(argument)
          print('Please try again')

      except:
          print('Error: I got some other error')
          print('Please try again')


# Test Case 1:  age = 10   (happy path; positive case)
# PASSED

# Test Case 2:  age = abc   (negative case)
# PASSED

# Test Case 3: age = -10 (negative case)
# PASSED

# Test Case 4: age = 150 (negative case)
# PASSED
```

```python
#@title Version 5: Take 'age' as input from the user (1 < age < 140)
# Given an option to the user to fix his/her mistake
# Make use of 'else' block for the try

# Handle the exceptions for incorrect inputs
class NegativeAgeException(Exception):
  pass

class TooBigAgeException(Exception):
  pass

# Keep on running this loop until we get the correct integer/age

while True:
      try:
          age = input('Enter your age: ')
          age = int(age)

          if (age < 0):
            raise NegativeAgeException('Error: Age can not be negative')

          if (age > 140):
            raise TooBigAgeException('Error: Age can not be more than 140')

      except ValueError:
          print('Value Error: Your input can not be converted to an integer.')
          print('Please try again')

      except NegativeAgeException as argument:
          print(argument)
          print('Please try again')


      except TooBigAgeException as argument:
          print(argument)
          print('Please try again')

      except:
          print('Error: I got some other error')
          print('Please try again')

      #Note: Typically whatever the code you have in else block
      #      can be put in the main 'try' block.
      #      This example is to demonstrate how the else block works
      else:
          age_2 = age + 5
          print('You are now', age, 'years old')
          print('You will be', age_2, 'after 5 years')
          break  #we are in happy path now



# Test Case 1:  age = 10   (happy path; positive case)
# PASSED

# Test Case 2:  age = abc   (negative case)
# PASSED

# Test Case 3: age = -10 (negative case)
# PASSED

# Test Case 4: age = 150 (negative case)
# PASSED
```

```python
#@title Version 6: Take 'age' as input from the user (1 < age < 140)
# Given an option to the user to fix his/her mistake
# Make use of 'else' block for the try
# Make use of 'finally' block

# Handle the exceptions for incorrect inputs
class NegativeAgeException(Exception):
  pass

class TooBigAgeException(Exception):
  pass

# Keep on running this loop until we get the correct integer/age
ERROR = True  # a tracker variable to run (or stop) the loop
while ERROR == True:
      try:
          age = input('Enter your age: ')
          age = int(age)

          if (age < 0):
            raise NegativeAgeException('Error: Age can not be negative')

          if (age > 140):
            raise TooBigAgeException('Error: Age can not be more than 140')

      except ValueError:
          print('Value Error: Your input can not be converted to an integer.')

      except NegativeAgeException as argument:
          print(argument)

      except TooBigAgeException as argument:
          print(argument)

      except:
          print('Error: I got some other error')

      #Note: Typically whatever the code you have in else block
      #      can be put in the main 'try' block.
      #      This example is to demonstrate how the else block works
      else:
          age_2 = age + 5
          print('You are now', age, 'years old')
          print('You will be', age_2, 'after 5 years')
          ERROR = False  #to break out of the loop

      # this code block will be run all the times
      finally:
        if ERROR:
          print('Please try again')



# Test Case 1:  age = 10   (happy path; positive case)
# PASSED

# Test Case 2:  age = abc   (negative case)
# PASSED

# Test Case 3: age = -10 (negative case)
# PASSED

# Test Case 4: age = 150 (negative case)
# PASSED
```

```python
#@title Handling Exceptions (basic construct)

INPUT_FILE = "quotes.txt"

try:
  with open(INPUT_FILE) as file_obj:
    text = file_obj.read()
    print(text)

except:
  print('Error: File not found')

```

```python
#@title Handling Exceptions (else and finally)

# 1. Run the code in try block
# 2. If everything in try is good, skip the except.
# 3. If there is else block, run it if try is good
# 4. If there is finally block, run it regardless (if code in try block is run without error or if there is any error)

INPUT_FILE = "python_input_files/sub1/names.txt"

try:
  with open(INPUT_FILE) as file_obj:
    text = file_obj.read()
    print('1. I read the file correctly')

except:
  print('2. Error: File not found')

else:
  print('3. I found the file; All is good')

finally:
  print('4. Whether things go wrong or right, I am going to run all the times')
```

```python
#@title You can raise your own custom exceptions
class NegativeAgeException(Exception):
  pass

class TooBigNumberException(Exception):
  pass



try:
  x = int(input('Give me your age and I will give you its reciprocal:'))
  if (x <= 0):
    raise NegativeAgeException('Error: The age can not be less than or equal to 0')
  if (x > 120):
    raise TooBigNumberException('Error: You can not input a number > 120')
  y = 1/x
  print('Here is the reciprocal of your age:', y)

except ZeroDivisionError:
  print('Error: You entered a zero')

except ValueError:
  print('Error: You need to input a number')

except NegativeAgeException:
  print('Error: You can not enter a number less than zero')

except TooBigNumberException:
  print('Error: You can not input a number greater than 120')

except:
  print('Error: Something has gone wrong')
```

```python
#@title BaseException and Exception

try:
    raise Exception

except BaseException:
    print("a")
except Exception:
    print("b")
except:
    print("c")

# BaseException is the common base class of all exceptions.
# One of its subclasses, Exception, is the base class of all the non-fatal exceptions.

x = issubclass(BaseException, Exception)
print(x)
y = issubclass(Exception, BaseException)
print(y)
```
