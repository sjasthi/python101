# 3.WHILE loop with break

````python
#@title WHILE loop to generate 10 random numbers (basic version)
#(where each number is beween 1 and 25)
import random

loop_counter = 1

while (loop_counter <= 10):
  random_x = random.randint(1, 25)
  print(loop_counter, "Random number: ", random_x)
  loop_counter += 1
````

````python
#@title while loop with break (I don't like 13) - Version 1
# I don't like 13! Stop the train!
# Note:  Execute this program multiple times
# Print 10 random numbers between 1 and 25.
# But if the random number generated is 13, just stop the loop

import random

# 1. initialization (starting point)
loop_counter = 1

# 2. condition
while (loop_counter <= 10):
  random_x = random.randint(1, 25)

  if random_x == 13:
    print("13 is the random number! I am stopping!")
    break   #just stop the loop; Exit the loop

  print(loop_counter, "Random number: ", random_x)

  # converging towards failure
  loop_counter = loop_counter + 1
````

````python
#@title while loop with break (I don't like 13) - Version 2
# I don't like 13! Stop the train!
# Note:  Execute this program multiple times
# Print 10 random numbers between 1 and 25.
# But if the random number generated is 13, just stop the loop

import random

broken_out_of_loop = False
loop_counter = 1

while (loop_counter <= 10):
  random_x = random.randint(1, 25)


  if random_x == 13:
    print("13 is the random number! I am stopping!")
    broken_out_of_loop = True
    break

  print(loop_counter, "Random number: ", random_x)
  loop_counter = loop_counter + 1


# print the status
if (broken_out_of_loop):
  print('while is exited due to break')
else:
  print('while loop is executed normally')
````

````python
#@title While loop with "continue"
# I don't like 13! Skip it and keep moving!
# Note:  Execute this program multiple times
# Print 10 random numbers between 1 and 25.
# But if the random number generated is 13, skip that number

import random

# 1. initialization (starting point)
loop_counter = 1

# 2. condition
while (loop_counter <= 10):
  random_x = random.randint(10, 15)

  if random_x == 13:
    print("13 is skipped! But I am moving ahead with the rest!")
    continue   # just ignore the rest of the code and go to the beginning of the loop

  print(loop_counter, "Random number: ", random_x)

  # 3. increment / decrement / how am I converging to failure?
  loop_counter = loop_counter + 1


````

````python
#@title While loop with sentinel
# How to setup WHILE loop if I do not know the number of students
# break, sentinel value and True are typically used together.
# Keep on running the loop (while True)
# Break out of the loop when some 'sentinel' value is reached


# 1. Initialize the variables
number_of_students = 0
total_age = 0

# 2. setup the while loop (sentinel is age = 0)
while True:
  # sentinel value is - any negative number
  age = input("Enter the age? (Must be a +ve number (Enter 0 or -1 to signal the end of inputs) ")
  age = int(age)

   # sentinel: if age is 0 or negative, break out of the loop
  if (age <= 0):
    break

  # if age is not 0, proceed in the normal path
  total_age = total_age + age
  number_of_students = number_of_students + 1

# find the average and display the output
average = total_age / number_of_students
print("Number of Students: ", number_of_students)
print("Total Age of Students: ", total_age)
print("Average age: ", average)
````

````python
#@title continue and break (an exercise)

'''
Run a While loop infinitely (using while True).

Inside the loop, generate two random numbers between (1 and 10)

If both the random numbers are equal, then the loop should skip those and continue with the next iteration.

If the sum of the random numbers is 10, then the loop should exit.

Otherwise, the loop will print the two random numbers and their sum.


For example, the typicaly output of the program can be as follows. (your output may be different)

Number 1: 8       Number 2: 3         Sum: 11
Number 1: 5       Number 2: 7         Sum: 12
Number 1: 4       Number 2: 6         Sum: 10
Exiting the program as the sum is 10

'''

# sentil loop with True condition
counter = 1
while True:
  num1 = random.randint(1, 10)
  num2 = random.randint(1, 10)

  if (num1 == num2):
    print("Both numbers are equal. I am continuing")
    continue

  if (num1 + num2 == 10):
    print("Sum of two numbers is 10: I am quitting")
    break

  print(counter, "--> No 1:", num1," No 2:", num1, "Sum: ", (num1 + num2))
  counter = counter + 1 # just a tracker variable to keep track of number of times the loop is run
````

````python
#@title While, True, break, continue - all pieces together
# Problem: Ask the user for a series of inputs. When the user signals the end, print the MAX, MIN and SUM

# prompt the user
print('Enter a series of +ve numbers (or (q)uit to stop):')
print('Note that ve and 0 will be ignored.')


# initialize the output variables
total = 0
min = -1
max = -1

# setup the while loop
while True:
    number = input('Enter the number: ')

    # break on the sentinel
    if (number == 'q'):
        break

    # convert str to int
    number = int(number)

    # ignore -ve and 0
    if (number <= 0):
        continue

    # update the total
    total = total + number

    # update max
    if (max == -1):
        max = number
    elif (number > max):
        max = number

    # update min
    if (min == -1):
        min = number
    elif (number < min):
        min = number


# print the outputs
print('Minimum = ', min)
print('Maximum = ', max)
print('Sum = ', total)
````

# 6.range() function --> Sequence

START (= 0), STOP, STEP ( = 1)



--> (English) can you list the numbers between 1 and 5?  2, 3, 4

---> (Python) RULE #1 in Range
---> First number is included, but not the last number
can you list the numbers between 1 and 5?  1, 2, 3, 4
can you list the numbers between -3 and 4?  -3, -2, -1, 0, 1, 2, 3

--> Python code
can you list the numbers between 1 and 5?   range(1, 5)
can you list the numbers between -3 and 4?  range(-3, 4)

---> (Python) RULE #1 in Range
---> if you don't specify the first number, it defaults to 0

can you list the numbers up to 7?  0, 1, 2, 3, 4, 5, 6

---> (Python code
---> if you don't specify the first number, it defaults to 0

--> (python code)
Can you list the numbers between 0 and 25 in the steps of 5? range(0, 25, 5)

# 9.Extending while loops with "else"

````python
#@title Extending 'while' loop: Happy Path
# If the loops runs normally, extend the loop and run the else block.
# if the loop ends abruptly (through break), do not extend the loop (do not run the else block)
n = 5
while n > 0:
    print(n)
    n = n-1
else:
   print('while loop executed all iterations without any interruption')
   print("SUCESS")
````

````python
#@title Extending 'while' loop: 'break' case
# If the loops runs normally, extend the loop and run the else block.
# if the loop ends abruptly (through break), do not extend the loop (do not run the else block)
n = 15
while n > 0:
    print(n)
    n = n-1
    if (n == 13):
       break
else:
   print('while loop executed all iterations without any interruption')
   print("SUCESS")
````

````python
#@title  Find the cube root using the while loops (using break)

'''
Assume you are given a positive integer variable named N.

 Write a piece of Python code that finds the cube root of N.

 The code prints the cube root if N is a perfect cube or it prints error if N is not a perfect cube.

 Hint: use a loop that increments a counter -- you decide when the counter should stop.
'''

N = int(input("Enter your number N:"))

counter = 1
# Keep on going until the result >= N
while True:
  result = counter * counter * counter

  if (result == N):
    print(counter)
    break
  elif (result > N):
    print("error")
    break
  else:
    counter = counter + 1

````
